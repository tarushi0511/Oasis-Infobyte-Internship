```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import plotly.express as px
```

```python
df=pd.read_csv("Unemp.csv")
```

```python
df.head()
```

```python
df.isnull()
```

```python
df.columns=['Region','Date','Frequency','Est Unemp Rate','Est Employed','Participation Rate','Area']
```

```python
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(10,8))
sns.heatmap(df.corr())
plt.show()
```

```python
df.columns=['Region','Date','Frequency','Est Unemp Rate','Est Employed','Participation Rate','Area']
plt.figure(figsize=(12, 10))
plt.title("Unemployment Record")
#plt.hist(df)
sns.histplot(x="Est Unemp Rate", hue="Region", data=df)
plt.show()
```

```python
plt.figure(figsize=(12, 10))
plt.title("Indian Unemployment")
sns.histplot(x="Est Unemp Rate", hue="Region", data=df)
plt.show()
```

```python
unemploment = df[["Region", "Est Unemp Rate"]]
figure = px.sunburst(unemploment, path=["Region", "Est Unemp Rate"],
                     values="Est Unemp Rate",
                     width=700, height=700, color_continuous_scale="RdY1Gn",
                     title="Unemployment Rate in India")
figure.show()
```
